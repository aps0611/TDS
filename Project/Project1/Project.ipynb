Import the libraries

```
from bs4 import BeautifulSoup as bs             # To pull Html and xml data
import requests                                 # to access the websites
import pandas as pd                             # for data manipulation
import numpy as np
```

Load the webpage

```
r = requests.get("https://www.senate.gov/legislative/LIS/roll_call_votes/vote1171/vote_117_1_00013.htm#state")
# if r is 200 the our request is succeeded 

# Create a new instance of beautiful soup
soup = bs(r.content)

# Print out the html:

contents = soup.prettify()
#print(contents)

#difficult to locate the tags: Use inspect
```

```
state_tags = soup.find_all(class_ = "contenttext responsive_col")
# print(state_tags)
```

```
my_list = []
for state in state_tags:
    my_list.append(state.text.strip())
print(my_list)
```

```
state_list = []
senator1_list = []
senator2_list = []

while my_list != []:
    state_list.append(my_list[0])
    my_list.pop(0)
    senator1_list.append(my_list[0])
    my_list.pop(0)
    senator2_list.append(my_list[0])
    my_list.pop(0)
```

```
df = pd.DataFrame(list(zip(state_list, senator1_list, senator2_list)),
               columns =['State', 'Senator1', 'Senator2'])
df.to_csv('senate.csv')
```
