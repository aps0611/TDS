### Task : Build a text report showing the main classification metrics.


```python
# import the library
from sklearn.metrics import classification_report
y_true = [0,1,2,2,2] # True label
y_pred = [0,0,2,2,1] # predicted label

# there are 3 classes in our example - 0,1,2
target_names = ['class 0','class 1', 'class 2']
```

```python
# get the matrix
print(classification_report(y_true, y_pred, target_names = target_names))
```

```python
# getting the confusion matrix

from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_true,y_pred)
mat

# Left: Actual # top predicted label
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# getting individual scores

print(precision_score(y_true, y_pred, average = None))
print(recall_score(y_true, y_pred, average = None))
print(f1_score(y_true, y_pred, average = None))
```
