### Making the changes to the data

```python
df.head()
```

```python
# Add all stats and create a total category so that we get to know which pokemon is best:

df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df.head(5)
```

```python
# Droping the specific column

df.drop(columns = ['Total'])
```

```python
## creating new colum total using iloc
# axis = 1 means adding horizontally
df['Total'] = df.iloc[:,4:10].sum(axis = 1)
df.head(5)
```

```python
cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df.head(5)
```

### Aggregate Statistics (Groupby)

```python
df = pd.read_csv('modified.csv')
df = df.drop(columns = ['Unnamed: 0', '#'])
df.head(5)
```

```python
df.groupby(['Type 1']).mean()
```

```python
# Who has best defense
df.groupby(['Type 1']).mean().sort_values('Defense', ascending = False)
```

```python
# who has best attack
df.groupby(['Type 1']).mean().sort_values('Attack', ascending = False)
```

```python
df.groupby(['Type 1']).sum()
```

```python
### counts

df.groupby(['Type 1']).count()['Name']
```

### Working with large amounts of data

```python
for df in pd.read_csv('modified.csv', chunksize = 5):
    print('Chunk df')
    print(df)
```

```python
new_df = pd.DataFrame(columns = df.columns)

for df in pd.read_csv('modified.csv', chunksize = 5):
    results = df.groupby(['Type 1']).count()
    
    new_df = pd.concat([new_df, results])
```
