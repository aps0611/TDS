Import the libraries

```
from bs4 import BeautifulSoup as bs
import requests #to access the websites
import pandas as pd
```

Load the webpage

```
r = requests.get("https://www.imdb.com/chart/top/")

# Convert to a beautiful Soup Object:
soup = bs(r.content)

# Print out the html:
contents = soup.prettify()
print(contents)

#difficult to locate the tags: Use inspect
```

```
# Create the empty list:

movie_title = []
movie_year = []
movie_rating = []
```

Extract HTML tag contents

```
imdb_table = soup.find(class_ = "chart full-width")
movie_titlecolumn = imdb_table.find_all(class_ = "titleColumn")
movie_ratingcolumn = imdb_table.find_all(class_ = "ratingColumn imdbRating")
```

```
for row in movie_titlecolumn:
    title = row.a.text  # row is inbuilt | a is tag name
    movie_title.append(title)
movie_title
```

```
for row in movie_titlecolumn:
    year = row.span.text
    movie_year.append(year)
movie_year
    
```

```
for row in movie_ratingcolumn:
    rating = row.strong.text #tag content extraction
    movie_rating.append(rating)
movie_rating
```

Create DataFrame

```
movie_df = pd.DataFrame({'Movie_Title': movie_title,'Movie_Year':movie_year,'Movie_Rating':movie_rating})
movie_df
```
