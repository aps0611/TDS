```
#import nominatim api
from geopy.geocoders import Nominatim
```

```
#activate nominatim geocoder
locator = Nominatim(user_agent="myGeocoder")
#type any address text:
location = locator.geocode("Clifton")
```

```
#Print Latitude and Longitude of that address:
print("Latitude = {}, Longitude = {}".format(location.latitude, location.longitude))
```

```
location.raw
```

Note: If you get a HTTP Error 403 from Nominatim, use your email ID instead of “myGeocoder” in Nominatim(user_agent="myGeocoder")
